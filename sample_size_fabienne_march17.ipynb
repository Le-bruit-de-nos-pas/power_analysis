```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats



alpha = 0.05  # Significance level (two-tailed test)
power = 0.80  # Statistical power

# Compute Z-scores for alpha and power
z_alpha_over_2 = stats.norm.ppf(1 - alpha / 2)  # Two-tailed test
z_beta = stats.norm.ppf(power)
            
# Fix observed proportion
p1_fixed = 0.40  

# Define range of expected proportions (p0) from 0.05 to 0.75 in steps of 0.05
p0_values = np.arange(0.05, 0.50, 0.05)
sample_sizes = []

# Compute required sample size for each p0
for p0 in p0_values:
    delta = abs(p1_fixed - p0)  # Absolute difference
    numerator = (z_alpha_over_2 + z_beta) ** 2 * (p0 * (1 - p0))
    denominator = delta ** 2
    n = np.ceil(numerator / denominator)  # Round up
    sample_sizes.append(n)

# Plot the results
plt.figure(figsize=(8, 6))
plt.plot(p0_values, sample_sizes, marker='o', linestyle='-', color='firebrick', markerfacecolor='none')
plt.xlabel("Expected Proportion (p0)")
plt.ylabel("Required Sample Size (n)")
plt.title("Sample Size Needed for Different Expected Proportions\n (Expected Population Prevalence = 40%)")
plt.grid(True)
plt.show()
```
